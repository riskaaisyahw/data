```python
rows = soup.find_all('div', class_='flex flex-col')
```

```python
data = []

for row in rows:
    merk = row.find(class_='flex flex-1 items-center truncate')
    if merk is not None:
        print(merk.text)
    else:
        print('-')
    keterangan = row.find(class_='overflow-hidden text-ellipsis whitespace-nowrap text-sm')
    if keterangan is not None:
        print(keterangan.text)
    else:
        print('-')
    harga_asli = row.find(class_='text-grey-60 line-through')
    if harga_asli is not None:
        print(harga_asli.text)
    else:
        print(tidak_diskon.text)
    diskon = row.find(class_='text-sm leading-6 text-grey-100')
    if diskon is not None:
        print(diskon.text)
    else:
        print('-')
    harga_diskon = row.find(class_='font-bold text-red-60')
    if harga_diskon is not None:
        print(harga_diskon.text)
    else:
        print('-')
    penawaran_lainnya = row.find(class_='truncate text-xxs font-bold tablet:text-xs')
    if penawaran_lainnya is not None:
        print(penawaran_lainnya.text)
    else:
        print('-')
    syarat_penawaran = row.find(class_='truncate text-xxxs leading-3 tablet:text-xxs')
    if syarat_penawaran is not None:
        print(syarat_penawaran.text)
    else:
        print('-')

    rich = {
        'merk' : merk,
        'keterangan' : keterangan,
        'harga_asli' : harga_asli,
        'diskon' : diskon,
        'harga_diskon': harga_diskon,
        'penawaran_lainnya' : penawaran_lainnya,
        'syarat_penawaran' : syarat_penawaran
    }
    
    data.append(rich)
```

```python
pd.DataFrame(data)
```
